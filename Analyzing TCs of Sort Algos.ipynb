# Analyzing the Time Complexities of Various Sorting Algorithms.

# Quick Sort

```python
import time
start_time =time.time()

import random
import time

def generate_random_numbers(n):

    numbers = []
    for i in range(n):
        numbers.append(random.randint(1, n))

    return numbers

def quickSort(array, low, high):

    if low >= high:
        return array

    pivot = array[high]

    i = low - 1 
    j = low      

    while j < high:
        if array[j] <= pivot:
            i += 1
            (array[i], array[j]) = (array[j], array[i])

        j += 1

    (array[i + 1], array[high]) = (array[high], array[i + 1])

    quickSort(array, low, i)
    quickSort(array, i + 2, high)

    return array

def main():

    for i in range(5):
        n = int(input("Enter the number of elements to be sorted: "))

        start_time = time.time()

        numbers = generate_random_numbers(n)

        numbers = quickSort(numbers, 0, len(numbers) - 1)

        end_time = time.time()

        print("Execution time for {} elements: {} seconds".format(n, round(end_time - start_time, 2)))

if __name__ == "__main__":
    main()
```

# Merge Sort

```python
import time
import random
def merge_sort(arr):

    if len(arr) <= 1:
        return arr
    
    mid = len(arr) // 2
    
    left_half = arr[:mid]
    right_half = arr[mid:]
    
    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)
    
    sorted_arr = merge(left_half, right_half)
    
    return sorted_arr

def merge(left, right):
    merged = []
    
    i = 0  
    j = 0 
    
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1
    
    while i < len(left):
        merged.append(left[i])
        i += 1
    
    while j < len(right):
        merged.append(right[j])
        j += 1
    
    return merged

import time

def calculate_time_complexity():

    for _ in range(5):
        size = int(input("Enter the input size: "))
        
        numbers = [random.randint(1, 1000) for _ in range(size)]
        
        start_time = time.time()
        
        sorted_numbers = merge_sort(numbers)
        
        end_time = time.time()
        
        elapsed_time = end_time - start_time
        
        print(f"Input size: {size} - Elapsed time: {elapsed_time} seconds")

calculate_time_complexity()

```

# Heap Sort

```python
import time
import random

def heapify(arr, n, i):
    largest = i  
    left = 2 * i + 1  
    right = 2 * i + 2  
    
    if left < n and arr[left] > arr[largest]:
        largest = left
    
    if right < n and arr[right] > arr[largest]:
        largest = right
    
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        
        heapify(arr, n, largest)
        
def heap_sort(arr):
    n = len(arr)
    
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)
    
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]  
        heapify(arr, i, 0)  
    
    return arr

def calculate_time_complexity():
    for _ in range(5):
        size = int(input("Enter the input size: "))
        
        numbers = [random.randint(1, 1000) for _ in range(size)]
        
        start_time = time.time()
        
        sorted_numbers = heap_sort(numbers)
        
        end_time = time.time()
        
        elapsed_time = end_time - start_time
        
        print(f"Input size: {size} - Elapsed time: {elapsed_time} seconds")

calculate_time_complexity()
```

# Insertion Sort

```python
import time

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

def main():
    for i in range(5):
        n = int(input("Enter the array size: "))
        start_time = time.time()
        arr = random.sample(range(1, n + 1), n)
        insertion_sort(arr)
        end_time = time.time()
        execution_time = end_time - start_time
        print("Array size:", n)
        print("Execution time:", "{:.10f}".format(execution_time), "seconds")

if __name__ == "__main__":
    main()
```
